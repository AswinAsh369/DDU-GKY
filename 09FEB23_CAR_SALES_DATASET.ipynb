## CAR SALES DATASET

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Car_sales.csv')
data.head(2)
```

```python
data.shape
```

```python
data.isnull().sum()
```

#### Drop na

```python
data.dropna(inplace=True)
data.info()
```

#### Checking the content in latest launch column

```python
data.Latest_Launch.unique()
```

#### Changing its data type inyo datetime

```python
data.Latest_Launch=pd.to_datetime(data.Latest_Launch)
```

```python
data.head(2)
```

```python
data.info()
```

#### We an see that it have changed to datetime 64[ns] format.

### Extracting year,month,data from the Latest_launch column

```python
data['Year']=pd.DatetimeIndex(data.Latest_Launch).year
```

```python
data['Month']=pd.DatetimeIndex(data.Latest_Launch).month
```

```python
data['Day']=pd.DatetimeIndex(data.Latest_Launch).day
```

```python
data.head(10)
```

### Convert data type of fuel efficiency into integer

```python
data.Fuel_efficiency=data.Fuel_efficiency.astype(int)
```

```python
data.info()
```

### Converting numerical column into categorical column

```python
data.Month.replace({
     1:'JANUARY',
     2:'FEBRUARY',
     3:'MARCH',
     4:'APRIL',
     5:'MAY',
     6:'JUNE',
     7:'JULY',
     8:'AUGUST',
     9:'SEPTEMBER',
     10:'OCTOBER',
     11:'NOVEMBER',
     12:'DECEMBER'
},inplace=True)
```
