### CREATING A DATAFRAME FROM A  LIST OF DICTIONARIES

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
name=['Arun','Varun','Leela','Geeta','Sanil','John']
age=[27,26,24,27,26,28]
```

```python
data=pd.DataFrame(list(zip(name,age)))
data
```

### Simplify the colunm name

```python
data=pd.DataFrame(list(zip(name,age)),columns=['Name','Age'])
```

## Filling a common value to all missing data

### Lets try filling 0 to all missing data

```python
data.fillna(0)
```

## Adding missing data to individual columns

#### The same method can be used to add missing data for various colunms differently. We just need to pass a dictionary as below.

```python
data.fillna({
    "Temperature":0,
    "Windspeed":5,
   "Status":"sunny" 
    })
```

### Forward Fill (row)

```python
data.fillna(method="ffill")
```

### Backward Fill (row)

```python
data.fillna(method="bfill")
```

```python
data.fillna(method="ffill",limit=10)
```
