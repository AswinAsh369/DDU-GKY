## PLOTING USING SEABORN WITH TITANIC DATASET

```python
data=pd.read_csv('./titanic.csv')
data.head()
```

# UNIVARIENT ANALYSIS

## KDE PLOT

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data.Survived)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data.Age)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data.Fare)
plt.show()
```

## HISTPLOT

```python
sns.histplot(data.Fare)
plt.show()
```

```python
sns.histplot(data.Age)
plt.show()
```

```python
data.head(1)
```

```python
sns.histplot(data.Class)
plt.show()
```

## BOX PLOT

```python
sns.boxplot('Age',data=data)
plt.show()
```

```python
sns.boxplot('Fare',data=data)
plt.show()
```

# MULTI VARIATE ANALYSIS

## LINE PLOT

```python
sns.lineplot('Age','Fare',data=data)
plt.title("Age vs Fare")
plt.show()
```

```python
data.head()
```

```python
sns.lineplot('Age','Survived',data=data)
plt.title("Age vs Survived")
plt.show()
```

```python
sns.lineplot('Fare','Survived',data=data)
plt.title("Fare vs Survived")
plt.show()
```

```python
sns.lineplot('Fare','Survived',data=data, hue='Sex')
plt.title("Fare vs Survived")
plt.show()
```

```python
sns.lineplot('Fare','Survived',data=data, hue='Pclass')
plt.title("Fare vs Survived")
plt.show()
```

## PIE CHART

```python
data.columns
```

```python
Pclass_Survived=data.groupby(['Pclass'])['Survived'].sum()
Pclass_Survived
```

```python
sns.set_style('ticks')
Pclass_Survived.plot.pie()
plt.legend()
plt.show()
```

```python
Pclass_Survived_sex_sum=data.groupby(['Pclass','Sex'])['Survived'].sum()
Pclass_Survived_sex_sum
```

```python
sns.set_style('ticks')
Pclass_Survived_sex_sum.plot.pie()
plt.legend()
plt.show()
```

```python
Fare_Survived_sex_sum=data.groupby(['Fare','Sex'])['Survived'].sum()
Pclass_Survived_sex_sum
```

```python
sns.set_style('ticks')
Fare_Survived_sex_sum.plot.pie()
plt.legend()
plt.show()
```

```python
Pclass_Survived_sex_sum=data.groupby(['Sex'])['Fare'].sum()
Pclass_Survived_sex_sum
```

```python
Pclass_Survived_sex_sum.plot.pie(autopct ='%1.2f%%')
plt.legend()
plt.show()
```

```python
Pclass_sex_survived1=data.groupby(['Pclass','Sex'])['Survived'].sum()
Pclass_sex_survived1.reset_index()
```

```python
Pclass_sex_survived1.plot.pie(autopct='%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1),loc='upper left',borderaxespad=0)
plt.show()
```

```python
Pclass_sex_survived1.plot.pie(autopct='%1.2f%%')
plt.legend()
plt.show()
```

## BAR CHART

```python
plt.figure(figsize=(4,3))
sns.countplot('Sex',hue='Survived',data=data)
plt.show()
```

## COUNT PLOT

```python
plt.figure(figsize=(4,3))
sns.countplot('Survived',data=data)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.set_style('darkgrid')
sns.countplot('Pclass',hue='Survived',data=data)
plt.title('Pclass : Survived vs Dead')
plt.show()
```

## FACTOR PLOT

```python
sns.set_style('ticks')
sns.factorplot('Pclass','Survived',hue='Sex',data=data)
plt.show()
```

## VIOLIN PLOT

```python
sns.set_style('dark')
plt.figure(figsize=(6,5))
sns.violinplot('Sex','Age',hue='Survived',data=data)
plt.show()
```

## BAR PLOT

```python
plt.figure(figsize=(8,4))
sns.barplot('SibSp','Survived',data=data)
plt.title('SibSp & Survived')
plt.show()
```

## PAIR PLOT

```python
sns.pairplot(data=data)
plt.show()
```

## HEAT MAP

```python
heat_map=data.corr()
sns.heatmap(heat_map)
plt.show()
```
