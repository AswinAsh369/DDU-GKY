### PLOTING WITH PANDAS

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Car_sales.csv')
data.sample(10)
```

#### Dropping na values

```python
data.dropna(inplace=True)
data.info()
```

#### Ploting on numerical value columns

### HISTOGRAM

```python
data.Horsepower.plot.hist(figsize=(6,3),cmap='gist_rainbow_r');
```

### Box Plot

```python
data.Sales_in_thousands.plot.box(figsize=(4,3),cmap='gist_earth');
```

```python
data.Horsepower.plot.box(figsize=(4,3),cmap='Blues_r');
```

### Subplots

```python
data.plot.box(figsize=(22,4),subplots=True);
```

```python
data.plot.hist(figsize=(6,25),subplots=True);
```

### Line Charts

```python
data.sort_values(by='__year_resale_value').plot.line('__year_resale_value','Price_in_thousands',figsize=(6,3),cmap='gnuplot');
```

### Bar Plot

```python
data.Manufacturer.value_counts().plot.bar(figsize=(9,2),cmap='turbo_r');
```

```python
data.sort_values(by='Price_in_thousands', ascending=False)[:10].plot.bar('Manufacturer','Price_in_thousands',figsize=(6,3));
```

### Area Chart

```python
data.plot.area('Manufacturer','Price_in_thousands',figsize=(6,3),cmap='Pastel2');
```

### Horizontal bar chart

```python
data.sort_values(by='Price_in_thousands')[:5].plot.barh('Model','Power_perf_factor',figsize=(6,4));
```

### Scatter Plot

```python
data.sort_values(by='Price_in_thousands')[:5].plot.scatter('Latest_Launch','Price_in_thousands',figsize=(10,3));
```

```python
data.plot.box(figsize=(22,5),subplots=True);
```

```python
data.sort_values(by='Engine_size',ascending=False)[:10].Engine_size.plot.pie();
```

### KDE Plot

```python
data.plot.kde(figsize=(10,3.8));
```

### Subplots

```python
data.plot.kde(figsize=(12,10),subplots=True);
```

### Hexbin Plot

```python
data.plot.hexbin('Curb_weight','Fuel_capacity',gridsize=15,cmap='autumn');
```
