## GROUP-BY USING TITANIC DATA

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
data.info()
```

### NUMBER OF PASSENGERS IN EACH CLASS

```python
data.Pclass.value_counts()
```

## SINGLE GROUP SINGLE COLUMN

### Count of passengers survived from each class

```python
Pclass_survived=data.groupby(['Pclass'])['Survived']
```

```python
Pclass_survived.sum()
```

  ### Count of male and female passengers survived

```python
Pclass_survived1=data.groupby(['Sex'])['Survived']
```

```python
Pclass_survived1.sum()
```

## MULTIPLE COLUMN

### Total amount paid by each class who survived

```python
Pclass_survived_fare=data.groupby(['Pclass'])['Survived','Fare']
Pclass_survived_fare.sum()
```

## MULTIPLE GROUP

### Count of male and female survived in each class

```python
Pclass_sex_survived=data.groupby(['Pclass','Sex'])['Survived']
```

```python
Pclass_sex_survived.sum()
```

```python
Pclass_survived_sex=data.groupby(['Pclass','Sex'],as_index=False)['Survived']
Pclass_survived_sex.sum()
```

## MULTIPLE FUNCTION

### Mean and sum of pclass survived

```python
Pclass_survived_mean_sum=data.groupby(['Pclass'])['Survived'].agg(['mean','sum'])
```

```python
Pclass_survived_mean_sum
```

```python
Pclass_survived_mean_sum_perc=data.groupby(['Pclass'])['Survived'].mean()*100
```

```python
Pclass_survived_mean_sum_perc
```

```python
data.to_csv('titanic_groupby.csv')
```
